## Why

::: {.incremental}
- Want to study chemical reactions in the quantum regime
  - Need few, very cold atoms confined in a small space.
  - At this point the atoms self-organise into a "Coulomb Crystal"

Tracking changes in the shape and bright points of the crystal shows how the numbers of each of the species are changing over time.

## Great - Where do we come in

Currently experimenters are having to compare the images taken from the microscope to simulation images to work out the number of each species.
